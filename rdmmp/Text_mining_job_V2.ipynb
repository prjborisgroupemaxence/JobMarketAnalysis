```python
from collections import defaultdict

'''Un `` RegexpTokenizer`` divise une chaîne en sous-chaînes en utilisant une expression régulière. 
Par exemple, le tokenizer suivant forme des jetons à partir de séquences alphabétiques, 
d'expressions monétaires et de toute autre séquence non-blanche:'''

tokenizer = nltk.RegexpTokenizer(r'\w+')

'''Elaboration d'une fonction pour extraire la fréquence des mots de chaque corpus par société'''

''' Initialiser les variables: 

corpora = defaultdict(list) => correspond au corpus. On lui affecte la fonction defaultdict()
stats, freq = dict(), dict() => stats: Correspond au nombre de mots // freq: à la fréquence des mots
corporabis = defaultdict(list) => D'autres variables sont intégrés pour distinguer les mots composés
statsbis, freqbis = dict(), dict()



'''

corpora = defaultdict(list)
stats, freq = dict(), dict()
corporabis = defaultdict(list)
statsbis, freqbis = dict(), dict()
    
def freq_stats_corpora(corpora, stats, freq, corporabis, statsbis, freqbis):
    
   
  

    # Création d'un corpus de tokens par company
    
    for company,posting in db.items():
        for posting in posting:
            tokens= tokenizer.tokenize(db[company]['Posting'])
            
            
           

            corpora[company] = tokens
    #Création d'un corpus pour les mots composés

            corporabis[company]= list(nltk.bigrams(tokens))
            

       
    # FreqDist() Fonction qui calcule la fréquence de mots
    # {'total': len(v)} calcule le nombre de mots dans le corpus. 

    for k, v in corpora.items():
        freq[k] = fq = dict(nltk.FreqDist(v))
        stats[k] = {'total': len(v)} 
        
    return (freq, stats, corpora)

    for k, v in corporabis.items():
        
        freqbis[k] = fq = dict(nltk.FreqDist(v))
        
        statsbis[k] = {'total': len(v)}
        
    return (freqbis, statsbis, corporabis)

    # Récupération des comptages
    # On va introduire dans une fonction, la fréquence, les stats et le corpus
    freq, stats, corpora = freq_stats_corpora()
    
    return freq, stats, corpora

    freqbis, statsbis, corporabis = freq_stats_corpora()
    
    return freqbis, statsbis, corporabis


    
        
    #Lancement de la fonction
    

freq_stats_corpora(corpora, stats, freq, freqbis, statsbis, corporabis)
  





```

```python
# Nettoyage du corpus

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords



sw = set()



def all_stop_words(sw):
    
'''
Pour cela, nous allons regrouper tous les mots vides c' est à dire les mots qui sont tellement communs qu'il est inutile 
de l'indexer ou de l'utiliser dans notre recherche
Création en dehors de la fonction d'un set vide
'''
    
    corpora = defaultdict(list)
    stats, freq = dict(), dict()
    corporabis = defaultdict(list)
    statsbis, freqbis = dict(), dict()
    
    # On appelle la fonction freq_stats_corpora() 
    
    freq_stats_corpora(corpora, stats, freq, corporabis, statsbis, freqbis)

      


    # Premièrement, on récupère la fréquence totale de chaque mot sur tout le corpus 

    # On instancie la variable feq_totale en lui affectant la fonction Counter de la librairie nltk. 
    #  La variable freq_totale récupère la fréquence des mots sans prendre en considération la company

    freq_totale = nltk.Counter()

    for k, v in corpora.items():
        freq_totale += freq[k]


    # Deuxièmement on décide manière un peu arbitraire du nombre de mots les plus fréquents à supprimer. On pourrait afficher un graphe d'évolution du nombre de mots pour se rendre compte et avoir une meilleure heuristique. 
    most_freq = list(zip(*freq_totale.most_common(30)))[0]

    # Troisièmement, nous allons regrouper les mots vides. 
    
    #On créé notre set de stopwords final qui cumule ainsi les 50 mots les plus fréquents du corpus ainsi que l'ensemble de stopwords par défaut présent dans la librairie NLTK
    
    
    
    # Nous introduisons dans la variable stop_words les mots vides qui appartiennent à la librairie NLTK


    stop_words = set(stopwords.words("french"))
    
    # On ajoute les mots les plus fréquents et les stop_words dans le set

    sw.update(most_freq)
    sw.update(tuple(stop_words))

    return sw

all_stop_words(sw)

    
```

```python
from collections import defaultdict

''' On reprocède à la tokenisation en prenant en supprimant les mots vides'''

'''Un `` RegexpTokenizer`` divise une chaîne en sous-chaînes en utilisant une expression régulière. 
Par exemple, le tokenizer suivant forme des jetons à partir de séquences alphabétiques, 
d'expressions monétaires et de toute autre séquence non-blanche:'''

tokenizer = nltk.RegexpTokenizer(r'\w+')

'''Elaboration d'une fonction pour extraire la fréquence des mots de chaque corpus par société'''

''' Initialiser les variables: 

corpora = defaultdict(list) => correspond au corpus. On lui affecte la fonction defaultdict()
stats, freq = dict(), dict() => stats: Correspond au nombre de mots // freq: à la fréquence des mots
corporabis = defaultdict(list) => D'autres variables sont intégrés pour distinguer les mots composés
statsbis, freqbis = dict(), dict() '''


corpora2 = defaultdict(list)
corpora2bis = defaultdict(list)
stats2, freq2 = dict(), dict()
stats2bis, freq2bis = dict(), dict()# On instancie corpora


def freq_stats_corpora2(freq2, stats2, corpora2, freq2bis, stats2bis, corpora2bis):
    
 
    
    

    # Nous allons supprimer du corpus les mots vides. Pour cela, nous appelons la fonction all-stop_words que nous venons de créer.
    sw = set()
    all_stop_words(sw)
    
    # Création d'un corpus de tokens par company en supprimant les mots vides
    
    for company,posting in db.items():
        for post in posting:
            tokens2= tokenizer.tokenize(db[company]['Posting'])
            

           
                  
    
    # On supprime tous les mots vides
        
            for w in tokens2:
                if not w in sw:
                    w = w.lower()
                    corpora2[company].append(w)   
                    
    # On effectue la séparation en tuple pour obtenir des couple de mots
            
            
            
            for w in tokens2:
                if not w in sw:
                    w = w.lower()
                    corpora2bis[company].append(w)      
            


    # FreqDist() Fonction qui calcule la fréquence de mots
    # {'total': len(v)} calcule le nombre de mots dans le corpus. 

    for k2, v2 in corpora2.items():        
        freq2[k2] = fq = nltk.FreqDist(v2)
        stats2[k2] = {'total': len(v2)} 
    return (freq2, stats2, corpora2)

    for k2, v2 in corpora2bis.items():
        v2 = list(nltk.bigrams(v2))
        freq2bis[k2] =  dict(nltk.FreqDist(v2))
        stats2bis[k2] = {'total': len(v2)} 
    return (freq2bis, stats2bis, corpora2bis)

    # Récupération des comptages
    # On va introduire dans une fonction, la fréquence, les stats et le corpus
    freq2, stats2, corpora2, freq2bis, stats2bis, corpora2bis = freq_stats_corpora2()
    return freq2, stats2, corpora2, freq2bis, stats2bis, corpora2bis

 
    
    
    

freq_stats_corpora2(freq2, stats2, corpora2, freq2bis, stats2bis, corpora2bis)
  





```

```python
#Merge des datasets"


languages = ['python', 'r', 'sql', 'java', 'c++', 'c#', 'scala', 'perl', 'julia', 
             'javascript', 'html', 'css', 'php', 'ruby', 'matlab', 'sas'] 

    
def my_merge(word_list):
    
    
    for word in word_list:     

        if word == tuple:

            dataframe = search_composite_word(word_list)


        else:

            dataframe = search_word(word_list)

    
    
      
    
    res = pd.merge(dataframe[0], dataframe[1], how="outer", left_index= False, right_index= False, on = 'Company')

    for i in range(2, len(word_list)): 

         res = pd.merge(res, dataframe[i], how="outer", left_index= False, right_index= False, on = 'Company')
            
    
    return res



df_languages = my_merge(languages)

df_languages
```
